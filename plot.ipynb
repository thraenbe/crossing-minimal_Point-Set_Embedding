```python
import matplotlib.pyplot as plt

x_values = []
y1_values = []
y2_values = []
y3_values = []
y4_values =[]

# Data
with open('/home/bela/uni-tuebingen/6/bachelor/code/build/cluster-E.txt', 'r') as file:
    for line in file:
        # Split line into parts, first part is x-value, remaining are y-values
        parts = line.split()
        x_values.append(float(parts[0].replace(":", "")) )
        y1_values.append(float(parts[1]))
        y2_values.append(float(parts[2]))
        y3_values.append(float(parts[3]))
        y4_values.append(float(parts[4]))

print(y4_values)

# Plotting
plt.plot(x_values, y1_values, marker='o', label=r'cr(iterative$(G_{W})$)')
plt.plot(x_values, y2_values, marker='o', label=r'$\frac{cr(\mathrm{iterative}(G_{W}))}{|V|}$')
plt.plot(x_values, y3_values, marker='o', label=r'cr(recursive$(G_{W})$)')
plt.plot(x_values, y4_values, marker='o', label=r'$\frac{cr(\mathrm{recursive}(G_{W}))}{|V|}$')


# Labels and Title
plt.xscale("log") 
plt.xlabel('P(e)')
plt.ylabel('crossings')
plt.title('')
plt.legend()
plt.grid(True)

# Display
plt.show()
```
